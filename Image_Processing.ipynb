PCA(Principal Component Analysis)

It is used for removing dimesions(extra) from the data

Using PCA either we can avoid the dimensions or we can simply merge them.

Using PCA, 3-D data can be converted to 2-D data 

So, here will be using a 3-color scale image(RGB image)

RGB- Red, Green, Blue
Red : [255,0,0]
Green : [0,255,0]
Blue : [0,0,255]

Here, will be image RGB image and then will reduce the image dimension to 2-D image(2-color scale image B/W)

```python
import numpy as np
from PIL import Image
```

```python
img = r"C:\Users\shrey\Downloads\WhatsApp Image 2026-07-21 at 4.00.12 PM.jpeg"
```

```python
img
```

```python
img = Image.open(img)
```

```python
img
```

```python
img.size
```

```python
img = img.resize((255,255))
```

```python
img
```

```python
i_a = np.array(img)
```

```python
i_a
```

```python
i_c = img.convert('L')
```

```python
i_c
```

```python
i_a1 = np.array(i_c)
```

```python
i_a1
```

```python
from sklearn.decomposition import PCA
```

```python
p = PCA(n_components=50)
```

```python
p
```

```python
p1 = p.fit_transform(i_a1)
```

```python
p1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(p1)
plt.show()
```

```python
p2 = p.inverse_transform(p1)
```

```python
p2
```

```python
plt.imshow(p2)
plt.show()
```
